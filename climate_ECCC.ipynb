```python
import pandas as pd
import os

# List of station IDs
station_ids = [
    27865, 10244, 10197, 3920, 3932, 10220, 3966, 30455, 10899, 3987, 27674, 7582,
    30682, 4057, 4061, 26912, 4121, 4140, 30435, 7633, 42967, 6901, 4243, 4255, 4268,
    10903, 4287, 27534, 41738, 26773, 4333, 30578, 4352, 4377, 6900, 43540, 4432,
    10911, 43046, 6904, 32128, 4509, 42183, 4525, 7733, 27604, 7747, 4607, 27528,
    30266, 4647, 9026, 4656, 4671, 7790, 32473, 4689, 4699, 31367, 4715, 9005, 27533,
    41983, 4760, 4761, 10999, 7844, 4816, 27647, 4835, 4859, 7868, 4898, 7870, 4905,
    4923, 27529, 7925, 31688, 5126, 26953, 31010, 26799
]

# Years of interest
years = [2004, 2016, 2024]

# Output dictionary to store averages
averages = {}

# Loop through station IDs and years
for station_id in station_ids:
    for year in years:
        file_name = f"climate_data_{station_id}_{year}_yearly.csv"
        
        # Check if the file exists in the current directory
        if os.path.exists(file_name):
            # Load the CSV file into a DataFrame
            df = pd.read_csv(file_name)
            
            # Compute column averages and store in the dictionary
            # Skip non-numeric columns
            numeric_cols = df.select_dtypes(include='number')
            avg_values = numeric_cols.mean()
            
            # Store results in dictionary
            averages[f"{station_id}_{year}"] = avg_values.to_dict()
        else:
            print(f"File not found: {file_name}")

# Convert the dictionary to a DataFrame for easier viewing
averages_df = pd.DataFrame.from_dict(averages, orient='index')

# Save the averages to a new CSV file
averages_df.to_csv("averages_climate_data.csv")

# Print the DataFrame
print(averages_df)
```
